Dado un tablero de N x N, se han de poder poner N reinas en el tablero

Definir variables

```python
import random



ROWS = 200
COLS = ROWS
QUEENS = ROWS
MUTATION_PCT = 0.1
POPULATION_SIZE = 100
SOLUTION = QUEENS


```

Crear should_mutate, mutate, select fitness y

```python
def merge(population, newPopulation):
  
  merged_population = population + newPopulation
  return merged_population

def create_agent(rows): # el gen
  not_suffled_agent = list(range(0, rows))
  return random.sample(not_suffled_agent, rows)

def create_population(population_size, rows):
   return [create_agent(rows) for x in range(population_size)]


def should_mutate(percetage):
  return random.random() <= percetage


def clean_repeated_agents(population):
  return list(map(list, set(map(tuple, population))))


def mutate(agent):

  if should_mutate(MUTATION_PCT):
    random_pos_2 = random.randint(0,ROWS - 1)
    random_pos_1 = random.randint(0,ROWS - 1)


    temp_value = agent[random_pos_1]
    agent[random_pos_1] = agent[random_pos_2]
    agent[random_pos_2] = temp_value

  return agent

def select_best_population(population, population_size, rows):
  clean_population = clean_repeated_agents(population)

  if (len(clean_population) < population_size):
    pending_agents = population_size - len(clean_population)
    clean_population = merge(create_population(pending_agents, rows), clean_population)



  scores = list(map(fitness, clean_population))
  ordered_scores, ordered_population = zip(*(sorted(zip(scores, clean_population))))
  scores_list, ordered_population_list = list(ordered_scores), list(ordered_population)

  return ordered_population_list[:population_size], scores_list[:population_size]

def count_horizontal_collisions(agent):
  collisions = 0

  for index, gen in enumerate(agent):
    temp_arr = agent.copy()
    temp_arr.pop(index)

    if (gen in temp_arr):
      collisions += 1

  return collisions

#devolverá el número de colisiones con la reina contadose así misma  abs( x1 - x2) == abs(y1, y2) 
def fitness(agent):
  queens_coords = list(map(lambda x: (x[0], x[1]), enumerate(agent)))
  collisions = 0
  collisions += count_horizontal_collisions(agent)

  for coord1 in queens_coords:
    for coord2 in queens_coords:
      x1, y1 = coord1
      x2, y2 = coord2

      if (abs(x1 - x2) == abs(y1 - y2)):
        collisions += 1
  return collisions
```

Definir la función para generar los gene, la población, seleccionar los mejorar candidatos,el ,y el crossover (reproducción)

```python
def crossover(agent1, agent2):
   half = len(agent1) // 2
   half_agent_1 = agent1[:half]
   half_agent_2 = agent2[:half]
   mixed_agent_1 = half_agent_2 + [x for x in agent1 if x not in half_agent_2]
   mixed_agent_2 = half_agent_1 + [x for x in agent2 if x not in half_agent_1]

   return mutate(mixed_agent_1), mutate(mixed_agent_2)
```

Método main

```python

def main(rows=ROWS, population_size=POPULATION_SIZE):
  population, best_solution = create_population(population_size, rows), 0
  generation = 0
  best_solution = None
  best_agent = None
  counter = 0
  while(SOLUTION != best_solution):
    population, scores = select_best_population(population, population_size, rows)
    generation += 1
    best_solution = scores[0]
    best_agent = population[0]
    counter += 1
    children_agents = []

    if (best_solution == SOLUTION):
      break

    random.shuffle(population)


    if generation % 100 == 0:
      print('generacion: ', generation, best_solution, best_agent)
      
    for i in range(0, len(population), 2):
      new_child_1, new_child_2 = crossover(population[i], population[i+1])
      
      children_agents = children_agents + [new_child_1, new_child_2]

    population[:] = merge(population, children_agents)

  print(f'La solución es {best_solution} y se ha encontrado en {generation} generaciones.')
  print(f'El mejor gen es: {best_agent}')

random.seed()
main(ROWS, POPULATION_SIZE)
```
